### **Visualization**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Sanjana\Downloads\project retrun\src\final_merged_dataset.csv")
df
```

```python
print(df.columns)
```

```python
target_col = 'heart_disease_present'
print("Target column:", target_col)
```

```python
plt.figure()
df[target_col].value_counts().plot(kind='bar')
plt.title('Heart Disease Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[num_cols].hist(figsize=(14,10))
plt.show()
```

```python
plt.figure()
plt.hist(df['age'], bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure()
sns.countplot(x='sex', hue=target_col, data=df)
plt.title('Gender vs Heart Disease Rate')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure()
sns.boxplot(x=target_col, y='serum_cholesterol_mg_per_dl', data=df)
plt.title('Cholesterol Boxplot')
plt.show()
```

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
plt.figure(figsize=(12,8))
sns.heatmap(df[num_cols].corr(), cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure()
sns.scatterplot(x='age', y='resting_blood_pressure', hue=target_col, data=df)
plt.title('Age vs Resting Blood Pressure')
plt.show()
```
